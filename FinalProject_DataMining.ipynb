## Creating Function to add Age Group Attribute

```python
def add_age_group(df):
  # make a column for age groups
  # 0 = 18-24
  # 1 = 25-35
  # 2 = 36-44
  # 3 = 45-55
  # 4 = 56-64
  # 5 = 65 - 100
  age_groups = []
  for i in range(len(df)):
    if df['Age'][i] >= 18 and df['Age'][i] <= 24:
      age_groups.append(0)
    elif df['Age'][i] >= 25 and df['Age'][i] <= 35:
      age_groups.append(1)
    elif df['Age'][i] >= 36 and df['Age'][i] <= 44:
      age_groups.append(2)
    elif df['Age'][i] >= 45 and df['Age'][i] <= 55:
      age_groups.append(3)
    elif  df['Age'][i] >= 56 and df['Age'][i] <= 64:
      age_groups.append(4)
    elif  df['Age'][i] >= 65 and df['Age'][i] <= 101:
      age_groups.append(5)
    else:
      age_groups.append(5)

  #new column : Age Group
  df['Age Group'] = age_groups

add_age_group(df_age)
print(df_age)
```

# Outliers
Look at the outliers in the insomnia boxplot above. For the people who have insomnia and who have a lot more sleep ( > 7 hours) than the majority of people who have insomina.

```python
# get max whiskers to use to find all the outliers
stats = df_health.groupby('Sleep Disorder')['Sleep Duration'].describe()
# get IQR for df_health
Q1 = stats.loc['Insomnia', '25%']
Q3 = stats.loc['Insomnia', '75%']
IQR = Q3 - Q1
max_whisker = Q3 + (IQR * 1.5)
```

# PCA, tSNE, UMAP for df_health: Sleep health and lifestyle dataset

```python
from sklearn.decomposition import PCA
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import TSNE
from umap import UMAP
```

```python
# drop columns not useful for PCA, tSNE, UMAP
df_encoded = df_health.drop(columns=['Person ID', 'Blood Pressure'])
df_encoded = pd.get_dummies(df_encoded, dtype=int)

#isolate the target / class labels
df_target = df_encoded['Has Sleep Disorder']
df_encoded = df_encoded.drop('Has Sleep Disorder', axis=1)

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

df_encoded['Daily Steps'] =scaler.fit_transform(df_encoded[['Daily Steps']])
df_encoded['Sleep Duration'] =scaler.fit_transform(df_encoded[['Sleep Duration']])
df_encoded['Physical Activity Level'] =scaler.fit_transform(df_encoded[['Physical Activity Level']])
df_encoded['Stress Level'] =scaler.fit_transform(df_encoded[['Stress Level']])
df_encoded['Heart Rate'] =scaler.fit_transform(df_encoded[['Heart Rate']])
df_encoded['Quality of Sleep'] =scaler.fit_transform(df_encoded[['Quality of Sleep']])
df_encoded['Age'] =scaler.fit_transform(df_encoded[['Age']])

# combine the columns and target together in one dataset
df_combined = pd.concat([df_encoded, df_target], axis=1)

# display(df_combined)
# display(df_encoded
# display(df_target)
```

### Fuction from PS3 to plot_mnist_scatter

```python
def plot_mnist_scatter(
        data,
        c,
        name,
        ax=None,
    ):
    if ax is None:
        ax = plt.gca()
    # cmap = plt.get_cmap('Spectral', 10)
    cmap = plt.get_cmap('berlin', 10)
    # x = data.iloc[:, 0]
    # y = data.iloc[:, 1]
    x, y = data[:, 0], data[:, 1]

    sc = ax.scatter(
        x=x, y=y, c=c, cmap=cmap,
        edgecolors='black', s=30,
        alpha=0.90,
    );
    plt.colorbar(sc);
    ax.set_aspect('equal', 'datalim');
    ax.set_title(name);

    return ax
```

```python
X = df_combined.iloc[:, :-1]   # all the rows, all columns except the last one
y = df_combined.iloc[:, -1].astype(int)    # just the last column with class labels
# Set the n_components=3
pca = PCA()   # keep all components
pca.fit(X)
pca_X = PCA(n_components=3).fit_transform(X)

fig, ax = plt.subplots(figsize=(10, 5));
plot_mnist_scatter(data=pca_X, c=y, name='PCA'); # Call the defined function
ax.set(
    xlabel='PCA_1',
    ylabel='PCA_2',
);
```

```python
pca_X = PCA(n_components=10).fit_transform(X)
tsne_X = TSNE(n_components=3, perplexity=30, max_iter=300, random_state=21).fit_transform(X)
umap_X = UMAP(n_components=3, random_state=21, n_neighbors=100).fit_transform(X)
fig, axs = plt.subplots(figsize=(20, 5), ncols=3);

plot_mnist_scatter(data=pca_X, c=y, name='PCA', ax=axs[0]);
plot_mnist_scatter(data=tsne_X, c=y, name='tSNE', ax=axs[1]);
plot_mnist_scatter(data=umap_X, c=y, name='UMAP', ax=axs[2]);
```

Note
- 1.0: pink points are Has Sleep Disorder is true
- 0.0: blue points are Has Sleep Disorder is false

conclusion:

The hope of implementing PCA, tSNE, and UMAP is to have 2 clusters, one for has a sleep disorder and another for does NOT have a sleep disorder.

From the PCA graph we can say is it not good at separating if a person has or has not a sleeping disorder. There is small cluster on the right bottom side of the graph of a points that does not has a sleeping disorder.

From the tSNE graph we can see more split clusters of about 3 has a sleep disorder. While there are about 3-6 clusters of does not has a sleeping disorder. tSNE does a better job of clustering than PCA but the clusters do overlap a lot and splint into multiple groups.

From the UMAP, a lot of points overlap and make tight clusters.  nothing can be discern.
